```python
###Best Strike Rate###
batter_stats = balls.groupby("Batter").agg({
    "BatsmanRun": "sum",
    "BallNumber": "count"
})

batter_stats["StrikeRate"] = (batter_stats["BatsmanRun"] / batter_stats["BallNumber"]) * 100

# filter players with at least 1000 runs
batter_stats = batter_stats[batter_stats["BatsmanRun"] >= 1000]

best_strike = batter_stats.sort_values("StrikeRate", ascending=False).head(10)

best_strike
```

```python
plt.figure(figsize=(10,5))

best_strike["StrikeRate"].plot(kind="bar")

plt.title("Best Strike Rate in IPL (Min 1000 Runs)")
plt.xlabel("Batter")
plt.ylabel("Strike Rate")

plt.show()
```

```python
###Runs Scored by each team###
team_runs = balls.groupby("BattingTeam")["TotalRun"].sum().sort_values(ascending=False)

team_runs
```

```python
plt.figure(figsize=(10,5))

team_runs.plot(kind="bar")

plt.title("Total Runs Scored by Each IPL Team")
plt.xlabel("Team")
plt.ylabel("Total Runs")

plt.show()
```

```python
###Most Economical Bowlers###
bowler_stats = balls.groupby("Bowler").agg({
    "TotalRun": "sum",
    "BallNumber": "count"
})

bowler_stats["Overs"] = bowler_stats["BallNumber"] / 6
bowler_stats["Economy"] = bowler_stats["TotalRun"] / bowler_stats["Overs"]

# filter bowlers with at least 300 balls
bowler_stats = bowler_stats[bowler_stats["BallNumber"] >= 300]

best_economy = bowler_stats.sort_values("Economy").head(10)

best_economy
```

```python
plt.figure(figsize=(10,5))

best_economy["Economy"].plot(kind="bar")

plt.title("Most Economical Bowlers in IPL")
plt.xlabel("Bowler")
plt.ylabel("Economy Rate")

plt.show()
```
